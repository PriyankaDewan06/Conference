```python
titanic_data.drop("Cabin", axis=1,inplace=True) # dropping unnecessary value
```

```python
titanic_data.dropna(inplace=True) # dropping na(not number value)
```

```python
sex =pd.get_dummies(titanic_data['Sex'],drop_first=True) # to convert catagorical variable for male & female(dummies)
sex.head(5)
```

```python
embark=pd.get_dummies(titanic_data["Embarked"],drop_first=True) #to convert catagorical variable for Q & S(dummies)
embark.head(5)
```

```python
Pcl=pd.get_dummies(titanic_data["Pclass"],drop_first=True) #to convert catagorical variable for passenger class 1,2,3(if both 2& 3 are zero then class is 1)
Pcl.head(5)
```

```python
titanic_data=pd.concat([titanic_data,sex,embark,Pcl],axis=1) #concatenate some values
```

```python
titanic_data.head(5)
```

```python
titanic_data.drop(['Sex','Embarked','PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
titanic_data.head(5)
```

```python
titanic_data.drop(['Pclass'],axis=1,inplace=True)
```

```python
titanic_data.head(5)
```

```python
titanic_data.drop(['Q'],axis=1,inplace=True)
```

```python
titanic_data.head(5)
```

```python
titanic_data.head(5)
```

**Train** **Data**

```python
X= titanic_data.drop("Survived",axis=1) # independent value 
y=titanic_data["Survived"]              # dependent value
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test) # making predictions
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions) #classification report
```

```python
from sklearn.metrics import confusion_matrix # to check accuracy 
```

```python
confusion_matrix(y_test,predictions) # confusion matrix= (8+32)/(8+9+6+32)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predictions)
```
